## Practice with digits dataset

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

data shows the image as an array of numerical values

```python
digits.data[0]
```

```python
plt.gray()
for i in digits.images[:5]:
    plt.matshow(i)
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=1, stratify=digits.target)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=2000) # had to increase max_iter because defualt iter wouldn't converge
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)

# use confusion matrix to check accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
